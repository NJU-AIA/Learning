```python
import sys, os
sys.path.append(os.path.abspath(".."))
from ddpm.data import get_loader
from ddpm.core import MyDDPM
from ddpm.train import training_loop
from ddpm.sampling import generate_new_images
from ddpm.visualize import show_images
from ddpm.models.attentionUnet import TinyUNet
from torch.optim import Adam
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

loader = get_loader(fashion=False, batch_size=128)
model = TinyUNet().to(device)
ddpm = MyDDPM(model, device=device)
```

```python
training_loop(ddpm, loader, n_epochs=20, optim=Adam(ddpm.parameters(), 0.001), device=device, store_path="saved_models/attentionUnet_mnist.pt")
```

```python
generated = generate_new_images(ddpm, n_samples=64, device=device, gif_name="gifs/attentionUnet_mnist.gif")
show_images(generated, "Final Results")
```
